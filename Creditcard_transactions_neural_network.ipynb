```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Dummy dataset (simple example)
# 6 input features → Fraud (0 or 1)
X = np.array([
    [100, 25, 5000, 1, 0, 1],
    [2000, 45, 20000, 0, 1, 0],
    [150, 30, 7000, 1, 0, 1],
    [3000, 50, 30000, 0, 1, 0]
])
y = np.array([1, 0, 1, 0])  # 1 = Fraud, 0 = Not Fraud
```

```python
# Build Neural Network
model = Sequential()

model.add(Dense(6, input_dim=6, activation='relu'))   # Input layer
model.add(Dense(4, activation='relu'))                # Hidden layer
model.add(Dense(1, activation='sigmoid'))             # Output layer
```

```python
# Compile model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Show model structure
model.summary()
```
